```python
from langchain.docstore.document import Document as LangchainDocument
from tqdm import tqdm  # Assuming tqdm is being used for a progress bar

def create_document_content(row):
    return (
        f"Policy Number: {row['Policy Number']}, "
        f"Age: {row['Age']}, "
        f"Gender: {row['Gender']}, "
        f"Health Conditions: {row['Health Conditions']}, "
        f"BMI: {row['BMI']}, "
        f"Occupation: {row['Occupation']}, "
        f"Geographic Location: {row['Geographic Location']}, "
        f"Annual Income: {row['Annual Income (INR)']}, "
        f"Family Medical History: {row['Family Medical History']}, "
        f"Coverage Amount: {row['Coverage Amount (INR)']}, "
        f"Deductible: {row['Deductible (INR)']}, "
        f"Claims History: {row['Claims History']}, "
        f"Marital Status: {row['Marital Status']}, "
        f"Exercise Frequency: {row['Exercise Frequency']}, "
        f"Dietary Habits: {row['Dietary Habits']}, "
        f"Physical Activity Level: {row['Physical Activity Level']}, "
        f"Medication Use: {row['Medication Use']}, "
        f"Hospitalization History: {row['Hospitalization History']}, "
        f"Premium: {row['Premium (INR)']}, "
        f"Grace Period: {row['Grace Period']}, "
        f"Exclusions: {row['Exclusions']}, "
        f"No Claim Bonus (NCB): {row['No Claim Bonus (NCB)']}, "
        f"Ambulance Charges: {row['Ambulance Charges (INR)']}, "
        f"Critical Illness Coverage: {row['Critical Illness Coverage (INR)']}, "
        f"Daycare Procedures Covered: {row['Daycare Procedures Covered']}, "
        f"Room Rent Limit: {row['Room Rent Limit (INR)']}"
    )

# Assuming `ds` is a list of dictionaries where each dictionary represents a row in your dataset
RAW_KNOWLEDGE_BASE = [LangchainDocument(page_content=create_document_content(doc)) for doc in tqdm(ds)]
```

```python

CSV_SEPARATORS = [
    "\n",
    ",",
    "\"",
    "",
]

```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from transformers import AutoTokenizer


EMBEDDING_MODEL_NAME = "thenlper/gte-small"

# Check internet connection before attempting to download
!ping -c 1 huggingface.co  # Test connection to huggingface.co

def split_documents(
    chunk_size: int,
    knowledge_base: List[LangchainDocument],
    tokenizer_name: Optional[str] = EMBEDDING_MODEL_NAME,
) -> List[LangchainDocument]:

    # Try to load the tokenizer, handling potential connection issues
    try:
        text_splitter = RecursiveCharacterTextSplitter.from_huggingface_tokenizer(
            AutoTokenizer.from_pretrained(tokenizer_name),
            chunk_size=chunk_size,
            chunk_overlap=int(chunk_size / 10),
            add_start_index=True,
            strip_whitespace=True,
            separators=CSV_SEPARATORS,
        )
    except OSError as e:
        print(f"Error loading tokenizer: {e}")
        print("Check your internet connection and try again.")
        return []  # Return an empty list if tokenizer loading fails

    docs_processed = []
    for doc in knowledge_base:
        docs_processed += text_splitter.split_documents([doc])

    unique_texts = {}
    docs_processed_unique = []
    for doc in docs_processed:
        if doc.page_content not in unique_texts:
            unique_texts[doc.page_content] = True
            docs_processed_unique.append(doc)

    return docs_processed_unique


docs_processed = split_documents(
    512,
    RAW_KNOWLEDGE_BASE,
    tokenizer_name=EMBEDDING_MODEL_NAME,
)

tokenizer = AutoTokenizer.from_pretrained(EMBEDDING_MODEL_NAME)
lengths = [len(tokenizer.encode(doc.page_content)) for doc in tqdm(docs_processed)]
fig = pd.Series(lengths).hist()
plt.title("Distribution of document lengths in the knowledge base (in count of tokens)")
plt.show()
```

```python
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores.utils import DistanceStrategy

embedding_model = HuggingFaceEmbeddings(
    model_name=EMBEDDING_MODEL_NAME,
    multi_process=True,
    model_kwargs={"device": "cuda"},
    encode_kwargs={"normalize_embeddings": True},
)

KNOWLEDGE_VECTOR_DATABASE = FAISS.from_documents(
    docs_processed, embedding_model, distance_strategy=DistanceStrategy.COSINE
)
```

```python

from transformers import pipeline
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

READER_MODEL_NAME = "HuggingFaceH4/zephyr-7b-beta"

# Check if GPU is available
if torch.cuda.is_available():
    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.bfloat16,
    )
    model = AutoModelForCausalLM.from_pretrained(
        READER_MODEL_NAME, quantization_config=bnb_config
    )
else:
    # Load the model without quantization if no GPU is available
    model = AutoModelForCausalLM.from_pretrained(READER_MODEL_NAME)

tokenizer = AutoTokenizer.from_pretrained(READER_MODEL_NAME)

READER_LLM = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    do_sample=True,
    temperature=0.2,
    repetition_penalty=1.1,
    return_full_text=False,
    max_new_tokens=500,
)
```

```python
prompt_in_chat_format = [
    {
        "role": "system",
        "content": """Using the information contained in the context,
give a comprehensive answer to the question.
Respond only to the question asked, response should be concise and relevant to the question.
Provide the number of the source document when relevant.
If the answer cannot be deduced from the context, do not give an answer.""",
    },
    {
        "role": "user",
        "content": """Context:
{context}
---
Now here is the question you need to answer.

You are supposed to take this data in and your main function is to give an approximitate
on how much insurance charge it would cost.
GIVE ANSWER IN RUPPEES ONLY. NOT IN DOLLARS
ONLY TELL THE PRICE, NOTHING ELSE

Question: {question}""",
    },
]
RAG_PROMPT_TEMPLATE = tokenizer.apply_chat_template(
    prompt_in_chat_format, tokenize=False, add_generation_prompt=True
)
# print(RAG_PROMPT_TEMPLATE)
```

```python
from ragatouille import RAGPretrainedModel

RERANKER = RAGPretrainedModel.from_pretrained("colbert-ir/colbertv2.0")
```

```python
from transformers import Pipeline


def answer_with_rag(
    question: str,
    llm: Pipeline,
    knowledge_index: FAISS,
    reranker: Optional[RAGPretrainedModel] = None,
    num_retrieved_docs: int = 30,
    num_docs_final: int = 5,
) -> Tuple[str, List[LangchainDocument]]:
    print("=> Retrieving documents...")
    relevant_docs = knowledge_index.similarity_search(
        query=question, k=num_retrieved_docs
    )
    relevant_docs = [doc.page_content for doc in relevant_docs]  # Keep only the text

    if reranker:
        print("=> Reranking documents...")
        relevant_docs = reranker.rerank(question, relevant_docs, k=num_docs_final)
        relevant_docs = [doc["content"] for doc in relevant_docs]

    relevant_docs = relevant_docs[:num_docs_final]

    context = "\nExtracted documents:\n"
    context += "".join(
        [f"Document {str(i)}:::\n" + doc for i, doc in enumerate(relevant_docs)]
    )

    final_prompt = RAG_PROMPT_TEMPLATE.format(question=question, context=context)

    print("=> Generating answer...")
    answer = llm(final_prompt)[0]["generated_text"]

    return answer, relevant_docs
```

```python
def calculate_bmi():
    print("You need to provide your weight and height to calculate BMI.")
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def get_user_info():
    print("Welcome to Insure Genie, Choose one of the following:")
    print("1. I want to enquire about an Insurance plan")
    print("2. I want general enquiry")
    choice = int(input("Enter the number corresponding to your choice: "))

    if choice == 2:
        general_enquiry = input("Thank you for choosing general enquiry. How can we assist you today? ")
        return general_enquiry

    age = input("Enter your age: ")
    gender = input("Enter your gender (Male/Female): ")

    diseases = ["None", "Thyroid Disorder", "Asthma", "Hypertension", "Diabetes", "Other"]
    print("Select a disease from the following options:")
    for i, disease in enumerate(diseases, 1):
        print(f"{i}. {disease}")

    disease_choice = int(input("Enter the number corresponding to your choice: "))
    if disease_choice == len(diseases):
        disease = input("Enter your disease: ")
    else:
        disease = diseases[disease_choice - 1]

    bmi_choice = input("Do you know your BMI? (Yes/No): ")
    if bmi_choice.lower() == 'yes':
        bmi = float(input("Enter your BMI: "))
    else:
        bmi = calculate_bmi()

    exercise_options = [
        "Active",
        "High",
        "Moderate",
        "Low"
    ]
    print("How often do you exercise?")
    for i, option in enumerate(exercise_options, 1):
        print(f"{i}. {option}")

    exercise_choice = int(input("Enter the number corresponding to your choice: "))
    exercise_frequency = exercise_options[exercise_choice - 1]

    user_info = (
        f"Age: {age}, Gender: {gender}, Health Condition: {disease}, "
        f"BMI: {bmi}, Exercise Frequency: {exercise_frequency}"
    )

    print("\nCollected User Information:")
    print(user_info)

    return user_info

user_info = get_user_info()
print(f"Output: {user_info}")
```

```python
def get_insurance_info(answer):
    import re

    patterns = {
        "Premium": r"Premium:\s*([\d,]+)",
        "Grace Period": r"Grace Period:\s*([\w\s]+)",
        "Exclusions": r"Exclusions:\s*([\w\s]+)",
        "No Claim Bonus (NCB)": r"No Claim Bonus \(NCB\):\s*([\d%]+)",
        "Ambulance Charges": r"Ambulance Charges:\s*([\d,]+)",
        "Critical Illness Charges": r"Critical Illness Charges:\s*([\d,]+)"
    }

    extracted_info = {key: re.search(pattern, answer) for key, pattern in patterns.items()}
    extracted_info = {key: match.group(1).replace(',', '') if match else "Not found" for key, match in extracted_info.items()}

    print(f"Your Approx. Premium would be around {extracted_info['Premium']} Rupees.")

    user_input = input("Do you have any other question? (yes/no): ").strip().lower()

    if user_input == 'yes':
        questions = {
            1: "What are the exclusions of this insurance plan?",
            2: "How much ambulance charges it can cover?",
            3: "What is the critical illness coverage amount?",
            4: "What is the grace period?",
            5: "What is the No Claim Bonus (NCB)?"
        }

        print("\nHere are some sample questions you can ask:")
        for num, question in questions.items():
            print(f"{num}. {question}")

        try:
            choice = int(input("\nChoose a question by entering the number (1-5): "))
            if choice in questions:
                chosen_question = questions[choice]
                if chosen_question == "What are the exclusions of this insurance plan?":
                    answer_key = "Exclusions"
                elif chosen_question == "How much ambulance charges it can cover?":
                    answer_key = "Ambulance Charges"
                elif chosen_question == "What is the critical illness coverage amount?":
                    answer_key = "Critical Illness Charges"
                elif chosen_question == "What is the grace period?":
                    answer_key = "Grace Period"
                elif chosen_question == "What is the No Claim Bonus (NCB)?":
                    answer_key = "No Claim Bonus (NCB)"

                print(f"\nAnswer to '{chosen_question}': {extracted_info.get(answer_key, 'Information not available')}")
            else:
                print("Invalid choice. Please enter a number between 1 and 5.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    else:
        print("Thank you!")
```
